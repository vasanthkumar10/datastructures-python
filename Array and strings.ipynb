```python
# uncompress
# Write a function, uncompress, that takes in a string as an argument. 
# The input string will be formatted into multiple groups according to the following pattern:

# <number><char>

# for example, '2c' or '3a'.

def uncompress(s):
    nums = '0123456789'
    result = ''
    i = j = 0
    while j < len(s):
        if (s[j] in nums):
            j += 1
        else:
            num = int(s[i:j])
            result += s[j] * num
            j += 1
            i = j
    return result

# test cases
tests = []

test0 = {
    'input': {
        's': "2c3a1t"
    },
    'output': 'ccaaat'
}
tests.append(test0)

test1 = {
    'input': {
        's': "4s2b"
    },
    'output': 'ssssbb'
}
tests.append(test1)

test2 = {
    'input': {
        's': "2p1o5p"
    },
    'output': 'ppoppppp'
}
tests.append(test2)

test3 = {
    'input': {
        's': "3n12e2z"
    },
    'output': 'nnneeeeeeeeeeeezz'
}
tests.append(test3)

test4 = {
    'input': {
        's': "7y"
    },
    'output': 'yyyyyyy'
}
tests.append(test4)

for i, test in enumerate(tests):
    print(f"testcase - {i}: {uncompress(**test['input']) == test['output']}")
```

```python
# compress
# Write a function, compress, that takes in a string as an argument. The function should return a compressed version of the string where consecutive occurrences of the same characters are compressed into the number of occurrences followed by the character. Single character occurrences should not be changed.

# 'aaa' compresses to '3a'
# 'cc' compresses to '2c'
# 't' should remain as 't'

def compress(s):
    s += '!'
    result = ''
    i = j = 0
    while j < len(s):
        if(s[i] == s[j]): j += 1
        else:
            size = j - i
            if(size == 1): result += s[i]  
            else: result += str(size) + s[i]
            i = j
    return result
    
    
    
# test cases
tests = []

test0 = {
    'input': {
        's': "ccaaatsss"
    },
    'output': '2c3at3s'
}
tests.append(test0)

test1 = {
    'input': {
        's': "ssssbbz"
    },
    'output': '4s2bz'
}
tests.append(test1)

test2 = {
    'input': {
        's': "ppoppppp"
    },
    'output': '2po5p'
}
tests.append(test2)

test3 = {
    'input': {
        's': "nnneeeeeeeeeeeezz"
    },
    'output': '3n12e2z'
}
tests.append(test3)

test4 = {
    'input': {
        's': "yyyyyyy"
    },
    'output': '7y'
}
tests.append(test4)

for i, test in enumerate(tests):
    print(f"testcase - {i}: {compress(**test['input']) == test['output']}")
```

```python
# anagrams
# Write a function, anagrams, that takes in two strings as arguments. 
# The function should return a boolean indicating whether or not the strings are anagrams. 
# Anagrams are strings that contain the same characters, but in any order.

def anagrams(s1, s2):
  # return sorted(s1) == sorted(s2)
    return char_count(s1) == char_count(s2)
  

def char_count(s):
    char_map = {}
    for char in s:
        if(char in char_map): char_map[char] += 1
        else: char_map[char] = 1
    return char_map

# test cases
tests = []

test0 = {
    'input': {
        's1': "restful",
        's2': "fluster"
    },
    'output': True
}
tests.append(test0)

test1 = {
    'input': {
        's1': "cats",
        's2': "tocs"
    },
    'output': False
}
tests.append(test1)

test2 = {
    'input': {
        's1': "monkeyswrite",
        's2': "newyorktimes"
    },
    'output': True
}
tests.append(test2)

test3 = {
    'input': {
        's1': "abbc",
        's2': "aabc"
    },
    'output': False
}
tests.append(test3)

test4 = {
    'input': {
        's1': "pp",
        's2': "oo"
    },
    'output': False
}
tests.append(test4)

for i, test in enumerate(tests):
    print(f"testcase - {i}: {anagrams(**test['input']) == test['output']}")
```

```python
# most frequent char
# Write a function, most_frequent_char, that takes in a string as an argument. 
# The function should return the most frequent character of the string. 
# If there are ties, return the character that appears earlier in the string.

def most_frequent_char(s):
    if len(s) == 0: return -1
    char_map = {}
    for char in s:
        if char in char_map: char_map[char] += 1
        else: char_map[char] = 1
            
    # max value in dict
    max_char = max(char_map, key= lambda char: char_map[char])
    return max_char
    

# test cases
tests = []

test0 = {
    'input': {
        's': "bookeeper"
    },
    'output': 'e'
}
tests.append(test0)

test1 = {
    'input': {
        's': "david"
    },
    'output': 'd'
}
tests.append(test1)

test2 = {
    'input': {
        's': ""
    },
    'output': -1
}
tests.append(test2)

test3 = {
    'input': {
        's': "eleventennine"
    },
    'output': 'e'
}
tests.append(test3)

test4 = {
    'input': {
        's': "yyyppp"
    },
    'output': 'y'
}
tests.append(test4)

for i, test in enumerate(tests):
    print(f"testcase - {i}: {most_frequent_char(**test['input']) == test['output']}")
```
