```python
# max value
# Write a function, max_value, that takes in list of numbers as an argument. 
# The function should return the largest number in the list.

# Solve this without using any built-in list methods.

def max_value(nums):
  # max = nums[0]
  # for i in range(1, len(nums)):
  #   if(nums[i] > max):
  #     max = nums[i]
    if len(nums) == 0: return -1
    max = float('-inf')
    for num in nums:
        if num > max: max = num
    return max

# test cases
tests = []

test0 = {
    'input': {
        'nums': []
    },
    'output': -1
}
tests.append(test0)

test1 = {
    'input': {
        'nums': [4, 7, 2, 8, 10, 9]
    },
    'output': 10
}
tests.append(test1)

test2 = {
    'input': {
        'nums': [10, 5, 40, 40.3]
    },
    'output': 40.3
}
tests.append(test2)

test3 = {
    'input': {
        'nums': [-5, -2, -1, -11]
    },
    'output': -1
}
tests.append(test3)

test4 = {
    'input': {
        'nums': [42]
    },
    'output': 42
}
tests.append(test4)

for i, test in enumerate(tests):
    print(f"testcase - {i}: {max_value(**test['input']) == test['output']}")
```

```python
# is prime
# Write a function, is_prime, that takes in a number as an argument. 
# The function should return a boolean indicating whether or not the given number is prime.

# A prime number is a number that is only divisible by two distinct numbers: 1 and itself.

# For example, 7 is a prime because it is only divisible by 1 and 7. 
# For example, 6 is not a prime because it is divisible by 1, 2, 3, and 6.

from math import sqrt, ceil

def is_prime(num):
    if num < 2: return False
    else:
        for i in range(2, ceil(sqrt(num))):
            if(num % i == 0): 
                return False
        return True

# test cases
tests = []

test0 = {
    'input': {
        'num': 0
    },
    'output': False
}
tests.append(test0)

test1 = {
    'input': {
        'num': 1
    },
    'output': False
}
tests.append(test1)

test2 = {
    'input': {
        'num': -10
    },
    'output': False
}
tests.append(test2)

test3 = {
    'input': {
        'num': 5
    },
    'output': True
}
tests.append(test3)

test4 = {
    'input': {
        'num': 8
    },
    'output': False
}
tests.append(test4)

for i, test in enumerate(tests):
    print(f"testcase - {i}: {is_prime(**test['input']) == test['output']}")
```
